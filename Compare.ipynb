```python
# convert everything to lowercase
utilities = list(map(str.lower, utilities))
ci = list(map(str.lower, ci))
```

```python
# remove duplicates
utilities = list(dict.fromkeys(utilities))
ci = list(dict.fromkeys(ci))
```

```python
print(f'Number of utilities companies: {len(utilities)}')
print(f'Number of c&i companies: {len(ci)}')
```

```python
# tokenize
utilities = list(map(str.split, utilities))
ci = list(map(str.split, ci))
```

```python
# remove common tokens
def remove_common_tokens(companies):
    common = ['the', 'pty', 'pty.', 'ltd', 'ltd.', 'limited', 'as', '-']
    for company in companies:
        for token in common:
            while token in company:
                company.remove(token)
```

```python
remove_common_tokens(utilities)
remove_common_tokens(ci)
```

```python
utilities = list(map(' '.join, utilities))
ci = list(map(' '.join, ci))

utilities.sort()
ci.sort()
```

```python
# exact matching
exact_matches = []

for company in ci:
    if company in utilities:
        exact_matches.append(company)
```

```python
exact_matches
```

```python
# fuzzy matching + first token match
from fuzzywuzzy import fuzz
```

```python
fuzzy_matches = []

for c in ci:
    for u in utilities:
        score = fuzz.token_set_ratio(c, u)
        if c.split()[0] == u.split()[0]:
            score += 50
        if score > 90:
            print(f'{c} | {u} | {score}')
```
